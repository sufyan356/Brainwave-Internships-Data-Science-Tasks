### Summary of super store sales

  The dataset contains transactional records from a superstore, capturing details of customer orders, shipments, and sales. It consists of various attributes related to orders, products, customers, and geographical regions. Key features of the dataset include:

  `Order Information:` Each transaction is uniquely identified by an Order ID, with details about the Order Date, Ship Date, and the shipping method used (Ship Mode).

  `Customer Details:` The dataset includes information on individual customers through Customer ID and Customer Name, and groups customers into segments (Consumer, Corporate, or Home Office) to better understand purchasing patterns.

  `Geographical Information:` Orders are tracked by customer location, including Country, City, State, Postal Code, and Region, allowing for analysis of regional sales trends.

  `Product Information:` Products are identified by Product ID and described through attributes such as Product Name, Category, and Sub-Category, providing insight into what types of products are being sold.

  `Sales and Financials:` The dataset tracks financial aspects, with columns such as Sales, allowing for the analysis of revenue generation across different products, customer segments, and regions.





#### Column Description:

`Row ID:` A unique identifier for each row or record in the dataset.

`Order ID:` A unique identifier for each customer order.

`Order Date:` The date when the order was placed.

`Ship Date:` The date when the order was shipped to the customer.

`Ship Mode:` The mode of shipping (e.g., Standard, Express) used to deliver the order.

`Customer ID:` A unique identifier assigned to each customer.

`Customer Name:` The full name of the customer who placed the order.

`Segment:` The customer segment (e.g., Consumer, Corporate, or Home Office) the customer belongs to.

`Country:` The country where the order was placed or shipped.

`City:` The city where the customer is located.

#### Data Clean

````python
super_store_df_copy = super_store_df.copy()
super_store_df_copy.info()
````

````python
#COMPLETENESS ISSUES:
#code:
super_store_df_copy['Postal Code'] = super_store_df_copy['Postal Code'].fillna('00000')

#test:
super_store_df_copy[super_store_df_copy['Postal Code'].isnull()]
````

````python
#VALIDITY ISSUES:

#code:
super_store_df_copy['Order Date'] = pd.to_datetime(super_store_df_copy['Order Date'] , format="%d/%m/%Y")
super_store_df_copy['Ship Date'] = pd.to_datetime(super_store_df_copy['Ship Date'] , format="%d/%m/%Y")

#test
super_store_df_copy.info()
````

````python
#VALIDITY ISSUES:

#code:
super_store_df_copy['Postal Code'] = pd.to_numeric(super_store_df_copy['Postal Code'], errors='coerce').fillna(0).astype(int).astype(str)
super_store_df_copy['Postal Code'] = super_store_df_copy['Postal Code'].str.zfill(5)

#test
super_store_df_copy.sample(10)
````

````python
# ACCURACY:
#code:
super_store_df_copy = super_store_df_copy[super_store_df_copy['Row ID'] != 3407]

#test:
super_store_df_copy[super_store_df_copy['Row ID'] == 3407]

````

````python
#CONSISTENCY:

#code:
super_store_df_copy['Category'] = super_store_df_copy['Category'].astype('category')
super_store_df_copy['Sub-Category'] = super_store_df_copy['Sub-Category'].astype('category')

#test:
super_store_df_copy.info()
````

#### Data Visualization & Getting Analysis

````python
### What is the total sales amount for each category?
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

tot_sales_each_catg = super_store_df_copy.groupby('Category')['Sales'].sum().sort_values(ascending=False)
tot_sales_each_catg.plot(kind='bar', color=colors, title='Total Sales by Category', ylabel = 'Sales' ,figsize = (7,5) , edgecolor='black')
plt.xticks(rotation=0)
plt.show()
````

**RESULT:** Technologies category is high in sale while office supplies is less in sale

````python
### Which segment contributes the most to sales?
segment_wise_sales = super_store_df_copy.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
segment_wise_sales.plot(kind='pie',figsize = (5,5) , autopct = '%.2f%%' , shadow=True)
plt.show()
````

**RESULT:** consumer has highest 50.77% of sales while
Home Office has low 18.78% of sales

````python
### What is the distribution of sales over time (by year)?
super_store_df_copy['Year'] = super_store_df_copy['Order Date'].dt.year
yearly_sales = super_store_df_copy.groupby('Year')['Sales'].sum()

yearly_sales.plot(kind='line',marker='o',figsize = (7,5) , title='Yearly Sales' , ylabel = 'sales' , color = 'green')
plt.show()
````

**RESULT** after 2016 throughout sales increases

````python
### What is the average shipping delay (difference between order date and ship date) across different shipping modes?

super_store_df_copy['shipping_delay'] = (super_store_df_copy['Ship Date'] - super_store_df_copy['Order Date']).dt.days
avg_shipping_delay = super_store_df_copy.groupby('Ship Mode')['shipping_delay'].mean().sort_values(ascending=False)

avg_shipping_delay.plot(kind = 'bar', figsize=(6,4), edgecolor = 'black', color = '#34A853', title='Average Shipping Delay by Ship Mode', ylabel = 'shipping delay')
plt.show()
````

**RESULT** ```The average shipping delay across different shipping modes indicates that Standard Class has the most shipping delay, followed by Second Class, First Class, and Same Day with the least delay time```

````python
### Which states generate the most sales, and how does that compare with the number of orders?

most_sales_states = super_store_df_copy.groupby('State')['Sales'].sum().sort_values(ascending=False)
no_of_orders_per_state = super_store_df_copy['State'].value_counts().sort_values(ascending=False)

fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(15,10))
ax[0].barh(most_sales_states.index, most_sales_states.values, color='#BF360C')
ax[0].set_xlabel('Sales')
ax[0].set_ylabel('State')
ax[0].set_title('Most Sales by State')

ax[1].barh(no_of_orders_per_state.index, no_of_orders_per_state.values, color='#E8710A')
ax[1].set_xlabel('orders')
ax[1].set_ylabel('State')
ax[1].set_title('Most orders by State')
plt.tight_layout()
````

**RESULT** ```California generates the highest sales and the most orders among states, while North Dakota records the lowest sales and Wyoming has the fewest orders, highlighting a significant disparity in sales and order performance across the states.```

````python
### Which products are the top 10 highest-grossing products by sales?
import textwrap

#top 10 highest grossing products
highest_grossing_products = super_store_df_copy.groupby('Product Name')['Sales'].sum().nlargest(10)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

wrapped_labels = [textwrap.fill(name, width=50) for name in highest_grossing_products.index]
ax[0].set_yticklabels(wrapped_labels, rotation=0, ha='right')

# Bar chart for highest-grossing products
ax[0].barh(highest_grossing_products.index, highest_grossing_products.values, color='#988884', edgecolor='black')
ax[0].set_xlabel('Sales')
ax[0].set_ylabel('Product Name')
ax[0].set_title('Sales Distribution of Top 10 Products')

# Pie chart for the same data
ax[1].pie(highest_grossing_products.values, labels=highest_grossing_products.index, autopct='%1.1f%%', shadow=True)
ax[1].set_title('Sales Distribution of Top 10 Products')

# Adjust layout
plt.tight_layout()
plt.show()
````

**RESULT** ```Among the top 10 highest-grossing products by sales, the Cannon Image Class 2200 stands out with the highest sales, while the High-Speed Automatic Electric ranks as the lowest.```

````python
### How are sales distributed by region?
sales_by_region = super_store_df_copy.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar', title='Sales by Region' ,ylabel='sales', edgecolor='black', color='#EF9C52')
plt.show()
````

**RESULT** ```Sales are distributed by region as follows: the West has the highest sales, followed by the East, while the Central region surpasses the South.```

````python
### What are the monthly sales trends, and is there any seasonality in the data?

monthly_sales = super_store_df_copy.resample('M', on='Order Date')['Sales'].sum()
monthly_sales.plot(title='Monthly Sales Trends', ylabel='Sales')
plt.show()

````

````python
### What is the relationship between sales and shipping delay? Does higher delay affect sales performance?

#using matplotlib
# super_store_df_copy.plot(kind='scatter', x='shipping_delay', y='Sales', title='relationship between sales and shipping delay')
# plt.shoSw()


#using seaborn
# Create a scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='shipping_delay', y='Sales', data=super_store_df_copy, color='blue', marker='o', line_kws={"color": "red"})

# Set titles and labels
plt.title('Relationship Between Shipping Delay and Sales')
plt.xlabel('Shipping Delay (days)')
plt.ylabel('Sales (dollars)')

# Show the plot
plt.tight_layout()
plt.show()



````

**RESULT** ```Straight line shows there is no relationship between shipping delay and sales which means higher delay time does not affect sales performance```
